```python
# appending calculated values to dataframe

dfObj = pd.DataFrame(new_table_df, columns = ['School', 'Type' , 'Size', 'Budget', "Reading", "Reading_Avg", "Math", "Math_Avg"]) 

dfObj['Reading%'] =  (dfObj['Reading'] / dfObj['Size'] ) * 100
dfObj['Math%'] =  (dfObj['Math'] / dfObj['Size']) * 100
dfObj['budget/student'] = (dfObj['Budget'] / dfObj['Size'])
dfObj['composite'] = (dfObj['Reading%'] + dfObj['Math%'] / 2)
dfObj['efficiency'] = (dfObj['composite'] / dfObj['budget/student'])
```

```python
# By school size (small, medium, large), and by budget

# school size -> size, reading avg, math avg, reading/math%, composite
# small = <1800, medium 1800 < 3600, large > 3600
sorted_by_size_df = dfObj.sort_values(by="Size")
sorted_by_size_df

# configure the bins
bin_sizes = [0, 1800, 3600, 5000]
bin_labels = ["Small", "Medium", "Large"]

# execute the binning function (pd.cut())
school_size_binned = pd.cut(sorted_by_size_df['Size'], bin_sizes, labels=bin_labels)

school_size_binned
```

```python
sorted_by_size_df['School Size'] = school_size_binned

sorted_by_size_df
```

```python
# calculate math average
math_avg_by_size = sorted_by_size_df.groupby(['School Size']).mean()['Math_Avg']
#math_avg_by_size

math_pass_rate = sorted_by_size_df.groupby(['School Size']).mean()['Math%']

# calculate reading average
reading_avg_by_size = sorted_by_size_df.groupby(['School Size']).mean()['Reading_Avg']

reading_pass_rate = sorted_by_size_df.groupby(['School Size']).mean()['Reading%']

overall_pass_rate = sorted_by_size_df.groupby(['School Size']).mean()['composite']
overall_pass_rate
```

```python
# create result dataframe
school_size_results = pd.DataFrame({
    'Math Average': math_avg_by_size,
    'Math Pass Rate': math_pass_rate,
    'Reading Average': reading_avg_by_size,
    'Reading Pass Rate': reading_pass_rate,
    'Overall Pass Rate': overall_pass_rate,
})

school_size_results
```

```python
# By school budget (low, below avg, above avg, high), and by budget

# school size -> size, reading avg, math avg, reading/math%, composite
# low = <1000000, below_avg < 2000000, above_avg < 3000000, High < 4000000
sorted_by_budget_df = dfObj.sort_values(by="Budget")
sorted_by_budget_df

# configure the bins
bin_budget = [0, 1000000, 1500000, 3000000, 4000000]
bin_bud_labels = ["Low", "Below Avg", "Above Avg", "High"]

# execute the binning function (pd.cut())
school_bud_binned = pd.cut(sorted_by_budget_df['Budget'], bin_budget, labels=bin_bud_labels)

school_bud_binned
```

```python
sorted_by_budget_df['Budget Band'] = school_bud_binned

sorted_by_budget_df
```
